```python
import json
import matplotlib.pyplot as plt

# Path to the metrics.json file
file_path = 'output/metrics.json'

# Lists to store iteration numbers and total loss values
iterations = []
total_loss = []

# Read the JSON lines from the file
with open(file_path, 'r') as f:
    for line in f:
        data = json.loads(line)
        if 'iteration' in data and 'total_loss' in data:
            iterations.append(data['iteration'])
            total_loss.append(data['total_loss'])

# Plot the total loss over iterations
plt.figure(figsize=(7, 6))
plt.plot(iterations, total_loss, label='Total Loss',color ='red')
plt.xlabel('Iteration')
plt.ylabel('Total Loss')
plt.title('Total Loss over Iterations')
plt.legend()
plt.grid(True)
plt.show()
```
